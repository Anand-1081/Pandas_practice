```python

import numpy as np 
import pandas as pd 
```

```python

data = {
    'A': [1, 2, np.nan, 4, 5],
    'B': [1, 2, 3, 4, 5],
    'C': [1, 2, 3, np.nan, np.nan],
    'D': [1, np.nan, np.nan, np.nan, 5]
}
df = pd.DataFrame(data)
```

```python
df
```

```python
# finding Missing data
```

```python
df.isna()
```

```python
df.isna().sum()
```

```python
df.isna().any()
```

```python
# removing missing data
```

```python
df
```

```python
df.dropna()  # it drop data row wise 
```

```python
df.dropna(thresh= 3)  # 3 -- null values
```

```python
df.dropna(thresh=4)
```

```python
df.dropna(thresh=5)
```

```python
# filling the missing data
```

```python
df
```

```python
df.fillna(0)  #  have to use inplace true to apply in real data  
# it fill data column wise
```

```python
values = {'A':0 , 'B':100 , 'C':300 , 'D':400}
df.fillna(value = values)
```

```python
df
```

```python
df.fillna(df.mean())  # it fill na with the mean value of the same row non na value
```
